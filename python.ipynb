```python
df_age = pd.read_csv('data/서울특별시 광진구_연령별인구현황_20231231.csv', encoding = 'ansi')
# 데이터 출처 : https://www.data.go.kr/data/15052328/fileData.do
df_age
```

```python
age_lst = list(df_age['연령'])
```
