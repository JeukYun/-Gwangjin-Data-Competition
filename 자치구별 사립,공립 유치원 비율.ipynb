### 자치구별 사립, 공립 비율 확인

- 자치구별 유치원 개소수 확인

```python
df_new = df_all.groupby(['자치구']).count()[['주소']].reset_index()
df_new
```

- 공립/사립 개소수 확인

```python
df_new = df_all.groupby(['자치구','사립,공립']).count()[['주소']].reset_index()
df_new
```

- 공립/ 사립 비율 확인

```python
Gu = df_new['자치구'].unique()
ratio = []

for gu in Gu:
    a = df_new[(df_new['자치구'] == gu) & (df_new['사립,공립'] == '공립')]['주소'].values[0]
    b = df_new[(df_new['자치구'] == gu) & (df_new['사립,공립'] == '사립')]['주소'].values[0]

    ratio.append({'자치구': gu, '공립/사립': a / b })

print(ratio)
```

```python
df_ratio = pd.DataFrame(ratio)
df_ratio
```
