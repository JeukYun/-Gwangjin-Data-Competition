# 광진구 행정동별 교육비 지출순위 확인

```python
df_all = pd.read_csv('data/raw_data/서울시 상권분석서비스(소득소비-행정동).csv', encoding = 'ansi')
df_all.head()
```

광진구 행정동 추출을 위해 광진구 인구 데이터 불러오기

```python
df = pd.read_csv('data/new_data/광진구 연령대별 인구.csv', encoding = 'ansi')
df.head()
```

```python
df.columns[2:-1]
```

```python
# 소득,소비데이터 행정동 확인
sorted(df_all['행정동_코드_명'].unique())

# '중곡제1동' 이 아닌 '중곡1동'으로 되어있음
```

```python
dong_lst = ['중곡1동', '중곡2동', '중곡3동', '중곡4동', '능동', '구의1동', '구의2동', '구의3동', 
            '광장동', '자양1동', '자양2동', '자양3동', '자양4동', '화양동', '군자동']
```

```python
df_new = df_all[df_all['행정동_코드_명'].isin(dong_lst)]
df_new.head()
```

```python
df_new['행정동_코드_명'].unique()
# 광진구 15개 동 추출 확인
```

```python
df_sort = df_new.groupby(['기준_년분기_코드', '행정동_코드_명']).sum().reset_index() # 기준 년분기, 행정코드에 해당하는 값은 모두 1개씩이므로 sum해도 무관
df_sort
```

- 행정동별 인구수 추가

```python
pop = df.sum()[2:]
pop
```

```python
pop = pop.to_dict()
pop_new = {key.replace('제', ''): value for key, value in pop.items()}
pop_new
```

```python
df_sort['인구수'] = df_sort['행정동_코드_명'].map(pop_new)
df_sort.head()
```

```python
df_sort.to_csv('data/new_data/광진구 행정동별 소득,소비.csv', index = False)
```
