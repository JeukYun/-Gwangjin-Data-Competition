## 광진구 연령대별 인구(행정동별)

- 5세 단위

```python
df = pd.read_csv('data/raw_data/서울특별시 광진구_연령별인구현황_20231231.csv', encoding = 'ansi')
# 데이터 출처 : https://www.data.go.kr/data/15052328/fileData.do
df
```

```python
df.columns = df.columns.str.replace('제', '')
df.columns
```

```python
# 일단 80대까지만
df_age = df.iloc[:90]
df_age
```

```python
df_age['연령숫자'] = pd.to_numeric(df_age['연령'].str.replace('세', ''))
# df_age
# '연령숫자'를 5로 나눈 나머지를 제거하여 5의 배수로 그룹화
df_age['연령그룹'] = (df_age['연령숫자'] // 5) * 5
df_age
```

```python
# 각 연령 그룹에 해당하는 열의 값을 합산하여 새로운 행으로 추가
df_grouped = df_age.groupby('연령그룹').sum().reset_index()
df_grouped
```

```python
df_grouped['연령그룹'] = ['0~4세', '5~9세', '10~14세', '15~19세', '20~24세', '25~29세', '30~34세', '35~39세', '40~44세', '45~49세', '50~54세', '55~59세', '60~64세', '65~69세', '70~74세', '75~79세', '80~84세', '85~89세']
```

```python
df_grouped.iloc[:, :-1]
```

```python
df_grouped.to_csv('data/new_data/광진구 연령대별 인구(5세단위).csv', index = False)
```

- 10세 단위

```python
age_lst = list(df['연령'])
```

```python
df_age = df.T
df_age.columns = age_lst
```

```python
df_age = df_age.iloc[1:, :-1]
df_age.head()
```

```python
for i in range(10):
    start_age = i * 10
    end_age = (i + 1) * 10 - 1
    age_range = f'{start_age}~{end_age}세'
    df_age[age_range] = df_age.loc[:, f'{start_age}세':f'{end_age}세'].sum(axis=1)
```

```python
df_age.head()
```

```python
df_age.columns
```

```python
df_age = df_age[['0~9세', '10~19세', '20~29세', '30~39세', '40~49세', '50~59세', '60~69세', '70~79세', '80~89세', '90~99세']][:-1].T #마지막 행은 데이터기준일자
df_age
```

```python
df_age = df_age.reset_index().rename(columns = {'index':'연령그룹'})
df_age.to_csv('data/new_data/광진구 연령대별 인구(10세단위).csv', index = False)
```
