```python
extracted_addresses = []

for address in lst:
    # 공백을 기준으로 split하여 주소를 분리
    split_address = address.split()
    
    # 분리된 주소 중에서 4번째까지만 추출하여 다시 공백으로 연결하여 저장
    extracted_address = ' '.join(split_address[:4])
    
    # 추출된 주소를 결과 리스트에 추가
    extracted_addresses.append(extracted_address)

# 결과 출력
print(extracted_addresses)
```

```python
addr_lst = extracted_addresses
```

```python
lat = [] # y
lng = [] # x

for addr in addr_lst:
    query = addr

    endpoint = "https://naveropenapi.apigw.ntruss.com/map-geocode/v2/geocode"
    url = f"{endpoint}?query={query}"

    # 헤더
    headers = {
        "X-NCP-APIGW-API-KEY-ID": client_id,
        "X-NCP-APIGW-API-KEY": client_secret,
    }

    # 요청
    res = requests.get(url, headers=headers)
    data = res.json()

    try:
        lat_val = data['addresses'][0]['x']
        lng_val = data['addresses'][0]['y']
    except:
        lat_val = 'None'
        lng_val = 'None'

    lat.append(lat_val)
    lng.append(lng_val)
```

```python
naDF = pd.DataFrame({'lat':lat, 'lng':lng})
naDF
```
