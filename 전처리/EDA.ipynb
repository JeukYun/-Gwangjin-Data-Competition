```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('EDA용.csv')
df.head()
```

```python
df['2030대비 출생아수(%)'] = round((df['23년 출생인구'] / df['2030_상주인구수'])*100, 2)
df
```

## 자치구별 0~9세 최대,최소 행정동  

```python
df_age_lg = df.groupby('자치구').apply(lambda x: x.nlargest(1, '0~9세')).reset_index(drop=True)
df_age_lg.head()
```

```python
df_age_sm = df.groupby('자치구').apply(lambda x: x.nsmallest(1, '0~9세')).reset_index(drop=True)
df_age_sm.head()
```

```python
df.columns
```

```python
col_lst = ['유해업소수', '대형쇼핑몰수', '도서관수', '미술관 박물관 극장수', '의료시설수(병원,의원,보건소,보건지소)', 
           '공원 수', '경로당 수', '보육시설정원', '보육시설현원', '보육시설포화도',
           '횡단보도수', '육교수', 'CCTV수', '월평균소득금액', '평균면적(m^2)', '면적당 가격(만원/m^2)',
           '주유소,가스충전소 수', '0~9세', '2030_상주인구수', '2030대비 출생아수(%)', '공립', '사립', '공립+사립']
```

```python
for col in col_lst:
    mean_val = round(df_age_lg[col].mean(),2)
    print(f"{col} 평균 : {mean_val}")
```

```python
for col in col_lst:
    mean_val = round(df_age_sm[col].mean(),2)
    print(f"{col} 평균 : {mean_val}")
```

```python
df_age = pd.concat([df_age_lg[['자치구', '행정동']], df_age_sm[['행정동']]], axis=1)
df_age.columns = ['자치구', '행정동_0~9세 최대', '행정동_0~9세 최소']
df_age
```
