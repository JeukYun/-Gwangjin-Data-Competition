### 유치원 교사 수 추가

```python
data.head()
```

```python
data_kinder = pd.read_csv('../data/raw_data/서울시 유치원 직위 자격별 교직원 현황.csv', encoding='ansi')
data_kinder.head()
```

```python
data_kinder.info()
```

```python
data_kinder.describe()
```

```python
# data와 data_kinder순서, 주소명이 모두 동일함
sum(data['주소'] == data_kinder['주소'])
```

```python
data['보직교사수'] = data_kinder['보직교사수']
data['일반교사수'] = data_kinder['일반교사수']
data['특수교사수'] = data_kinder['특수교사수']
data['기간제교사수'] = data_kinder['기간제교사수']
data['총 교사수'] = data_kinder['보직교사수'] + data_kinder['일반교사수'] + data_kinder['특수교사수'] + data_kinder['기간제교사수']
data.head()
```

```python
data.to_csv('../data/raw_data/서울 유치원 전체(위경도추가).csv', index = False)
```
