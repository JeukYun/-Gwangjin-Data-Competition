```python
import pandas as pd
```

## 국공립 어린이집

```python
df1 = pd.read_csv('서울시 어린이집 정보(표준 데이터).csv')
df1.head()
```

```python
df1.columns
```

```python
df1 = df1[['시군구명', '어린이집유형', '어린이집명', '정원', '운영현황']]
df1.head()
```

```python
df1 = df1[(df1['운영현황'] == '정상') | (df1['운영현황'] == '재개')]
df1
```

```python
df_kdg = df1[df1['어린이집유형'] == '국공립']
```

```python
df_kdg
```

```python
df_kdg[df_kdg['정원'] == 0]
```

- 정원이 0인 곳이 101곳.. -> 이후 정원이 꼭 필요하면 자치구별 평균값으로 채우기

```python
df_kdg_size = df_kdg[['시군구명']].groupby('시군구명').size().to_frame().reset_index()
df_kdg_size
```

```python
df_kdg_size.columns = ['자치구', '공립어린이집수']
df_kdg_size
```

## 공립 유치원

```python
df2.info()
```

```python
df2.columns
```

## 사립 추가, 출생인구 추가

```python
df_kdg_private = df1[df1['어린이집유형'] == '민간']
df_kdg_private
```

```python
df_kdg_private_size = df_kdg_private.groupby('시군구명').size().to_frame().reset_index()
df_kdg_private_size.columns = ['자치구', '사립어린이집수']
df_kdg_private_size
```

```python
df_kdg2_private = df2[df2['공립/사립'] == '사립']
df_kdg2_private
```

```python
df_kdg2_private_size = df_kdg2_private.groupby('자치구').size().to_frame().reset_index()
df_kdg2_private_size.columns = ['자치구', '사립유치원수']
df_kdg2_private_size
```

```python
df3 = pd.read_csv('출생·사망(신고기준)(동별)_20240421164403.csv')
df3
```

```python
df3.drop(columns = '동별(1)', inplace = True)
df3
```

```python
df3 = df3[df3['동별(3)'] == '소계'].iloc[1:]
df3.reset_index(drop = True, inplace = True)
df3
```

```python
df3.columns
```

```python
col_lst = ['2023. 01', '2023. 02', '2023. 03', '2023. 04', '2023. 05', '2023. 06', 
           '2023. 07', '2023. 08', '2023. 09', '2023. 10', '2023. 11', '2023. 12']

for col in col_lst:
    df3[col] = df3[col].astype('int')

df3.info()
```

```python
df3['23년 출생인구'] = df3.sum(axis=1)
df3
```

```python
df_pop_size = df3[['동별(2)', '23년 출생인구']]
df_pop_size.columns = ['자치구', '23년 출생인구']
df_pop_size
```

```python
df_born = pd.read_csv('합계출산율+및+모의+연령별+출산율_20240428133549.csv')
```

```python
df_born = df_born[['자치구별(1)', '2022']].iloc[2:]
df_born.columns = ['자치구', '22년 출산율']
df_born
```

```python
df_born_size = df_born[df_born['자치구'] != '서울특별시'].reset_index(drop = True)
df_born_size
```

## 합치기

```python
df_new2 = pd.merge(pd.merge(pd.merge(df_kdg_private_size, df_kdg2_private_size, on='자치구'), df_pop_size, on = '자치구'), df_born_size, on = '자치구')
df_new2
```
